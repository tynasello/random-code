#### Polynomial regression using scikit-learn

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# x - nasal length (mm ¥10) of gray kangaroo
# y - nasal width (mm ¥ 10) of gray kangaroo

x = np.array([609, 629, 620, 564, 645, 493, 606, 660, 630, 672, 778,  616,  727,  810, 778, 823, 755, 710, 701, 803, 855, 838, 830,
             864, 635, 565, 562, 580, 596, 597, 636, 559, 615, 740, 677, 675, 629, 692, 710, 730, 763, 686, 717, 737, 816]).reshape((-1, 1))
y = np.array([241, 222, 233, 207, 247, 189, 226, 240, 215, 231, 263, 220, 271, 284, 279, 272, 268, 278, 238, 255, 308,
             281, 288, 306, 236, 204, 216, 225, 220, 219, 201, 213, 228, 234, 237, 217, 211, 238, 221, 281, 292, 251, 231, 275, 275])
```

```python
# get feature matrix for polynomial combinations of x with degree 2
poly = PolynomialFeatures(degree=2).fit_transform(x)
# looks like [[1, x, x^2], ...]
poly.shape
```

```python
# fit model
polynomialRegressionModel = LinearRegression().fit(poly, y)

# A coefficient of determination closer to 1 than 0 indicates the model can fit data well
c_d = polynomialRegressionModel.score(poly, y)
c_d
```

```python
domain = np.arange(math.floor(min(x)), math.ceil(max(x))+2).reshape((-1,1))
y_pred = polynomialRegressionModel.predict(PolynomialFeatures(degree=2).fit_transform(domain))

# x, y points
plt.scatter(x, y, s=5)
# polynomial regression line
plt.plot(domain, y_pred, linewidth=1)
plt.show()
```
