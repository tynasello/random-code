#### Simple single-variate linear regression model using SciPy

```python
from scipy import stats
import matplotlib.pyplot as plt
```

```python
x = [2,4,6,8,10,12]
y = [7,21,34,41,49,61]
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)

def f(x):
  return slope * x + intercept

y_pred = list(map(f, x))

plt.plot(x, y_pred, linewidth=1)
plt.scatter(x, y, s=5)
plt.show()
```

```python
# Coefficient of correlation. Shows there is a linear relationship between x and y.
r
```

#### Simple single-variate linear regression model using scikit-learn

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# x - chirps/sec for the striped ground cricket
# y - temperature in degrees Fahrenheit

x = np.array([20, 16, 19.79999924, 18.39999962, 17.10000038, 15.5, 14.69999981, 17.10000038,
             15.39999962, 16.20000076, 15, 17.20000076, 16, 17, 14.39999962]).reshape((-1, 1))

y = np.array([88.59999847, 71.59999847, 93.30000305, 84.30000305, 80.59999847, 75.19999695, 69.69999695,
             82, 69.40000153, 83.30000305, 79.59999847, 82.59999847, 80.59999847, 83.5, 76.30000305])
```

```python
linearRegessionModel = LinearRegression()
linearRegessionModel = linearRegessionModel.fit(x, y)
```

```python
# A coefficient of determination closer to 1 than 0 indicates the model is fitting data well
c_d = linearRegessionModel.score(x, y)
c_d
```

```python
domain = np.arange(math.floor(min(x)), math.ceil(max(x))+2).reshape((-1,1))
y_pred = linearRegessionModel.predict(domain)

# x, y points
plt.scatter(x, y, s=5)
# linear regression line
plt.plot(domain, y_pred, linewidth=1)
plt.show()
```
